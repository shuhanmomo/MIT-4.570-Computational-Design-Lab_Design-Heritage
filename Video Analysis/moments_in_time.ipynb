```python
from os import listdir
locations =['bostoncommon','kowloonpark','luxembourggardens']
location = locations[1]
videos = listdir(f'/content/drive/MyDrive/MIT design heritage/W3 video analysis/moments in time/input/{location}')
print(videos)
```

```python
videos_df = pd.DataFrame(videos)# create one-column pd dataframe
videos_df.columns = ['filename']
videos_df.head()
```

```python
def load_categories(filename):
    """Load categories."""
    with open(filename) as f:
        return [line.rstrip() for line in f.readlines()]

categories = load_categories('category_momentsv2.txt')
print(categories)
```

```python
for i in range(len(categories)):# will populate the picture has how much probability belonging to one class
    videos_df[categories[i]] = 0

videos_df.head()
```

```python
# fit the model
second_interval = 3

for index,video in enumerate(videos[1:51]):
  filename = video
  data = []
  timestamp = 0
  os.system("ffmpeg -i " + filename +f" -c copy -map 0 -segment_time 00:00:03 -f segment -reset_timestamps 1 process/{location}/{video[:-4]}-output%03d.mp4")   
  result = subprocess.run(["ffprobe", "-v", "error", "-show_entries",
        "format=duration", "-of",
        "default=noprint_wrappers=1:nokey=1", filename],
    stdout=subprocess.PIPE,
    stderr=subprocess.STDOUT)
  try:
    d = math.floor(float(result.stdout))
    segments = math.floor(d / second_interval)
    
    for i in range(0, segments):
      subdata = []
      if i <= 9:
          name = f"process/{location}/{video[:-4]}-output00" + str(i) + ".mp4"
      elif i >= 10 and i <= 99:
          name = f"process/{location}/{video[:-4]}-output0" + str(i) + ".mp4"
      else:
          name = f"process/{location}/{video[:-4]}-output" + str(i) + ".mp4"
      x = subprocess.Popen(["python", "test_video_test.py", "--video_file", name, "--arch", "resnet3d50"], stdout=subprocess.PIPE)
      (out, err) = x.communicate()
      subdata = out.decode('utf-8').split('\n')
      subdata.pop()
      subdata.insert(0,timestamp)
      timestamp += second_interval
      data.append(subdata)
      for k in range(1,len(subdata)):
        if k%2!=0:
          videos_df.loc[index, subdata[k]] += float(subdata[k+1])/segments
    print(f'processing:{index}/{len(videos)}')
  except Exception as e:
    print(f'processing:{index}/{len(videos)} but {e} occurs at {video}')



```
